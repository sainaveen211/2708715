```python
import pandas as pd

# Load the dataset
data = pd.read_csv("spotify_long_tracks_2014_2024.csv")

# Display the first few rows of the dataset
print(data.head())

# Summary statistics of the numerical variables
print(data.describe())

# Check for missing values
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data['Duration (Minutes)'] / (1000 * 60), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Track Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('spotify_long_tracks_2014_2024.csv')

# Select columns for scatter plot
duration_minutes = data['Duration (Minutes)']
artists = data['ID']

# Plot scatter plot
plt.figure(figsize=(26, 6))
plt.scatter(artists, duration_minutes, color='blue', alpha=0.5)
plt.title('Duration of Tracks by Artists')
plt.xlabel('Artists')
plt.ylabel('Duration (Minutes)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
